## Loading data from Compressed Files

In big data ere - it is likely to read data from compressed files.

One way to unzip the data is to use gzip. 

```python
import gzip 
import json

with gzip.open('data/poland-bankruptcy-data-2008.json.gz', 'rt') as f:
    poland_data = json.load(f)
```

```python
# Check the type of poland_data
# type(poland_data)

# Check the keys in the poland_data dictionary
poland_data.keys()
```

```python
# Using the `from_dict`  Pandas function to read the data

# DataFrame()
df = pd.DataFrame.from_dict(poland_data['data'])

df.head()
```

# Pickle File
A pickle file is a file that stores a Python object in a serialized binary format using the built-in Python module called `pickle`

Understand Pickling - Serialization and Deserializing.

`Serialization` means converting an object (like a list, dictionary, machine learning model, etc.) into bytes so it can be:
* saved to disk,
* transferred over a network,
* or loaded later.

A pickle file commonly uses extensions like:

* .pkl
* .pickle

`Deserializing` is the reverse process, turning a stream of bytes into an object in memory.

```python
# Example implementation using a dictionary with lists as values
fruits = {
    "apple": ["red", "green"],
    "banana": ["yellow", "green"],
    "grape": ["purple", "green"],
    "watermelon": ["green", "red"],
    "peach": ["pink", "yellow"],
    "pear": ["green", "yellow"],
    "pawpaw": ["green", "yellow"]
    
}

fruits
```

```python
import pickle 
import os


'''
"wb" means write binary
pickle.dump() writes the object into the file
'''

# create folder if it doesn't exist
os.makedirs('./pickle-files', exist_ok=True)

pickle.dump(fruits, open('./pickle-files/fruits.pkl', 'wb'))

```

The `pickle-files` inside this directory `notebooks/python-basics/pickle-files` contains a file named `fruits.pkl`

```python
# We can read the pickled file back into a Python object using pickle.load()

with open('./pickle-files/fruits.pkl', "rb") as f:
    unpickled_fruits = pickle.load(f)

# Call the unpickled_fruits variable to see the contents
unpickled_fruits
```
